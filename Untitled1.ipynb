```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random
from IPython.display import Image, display
import numpy as np

# üì• Entrada de datos del paciente
nombre = input("Nombre del paciente: ")
edad = int(input("Edad del paciente: "))
spo2 = float(input("SpO‚ÇÇ inicial (%): "))
tiempo_simulacion = int(input("Duraci√≥n de la simulaci√≥n (segundos): "))
spo2_critico = float(input("SpO‚ÇÇ cr√≠tico m√≠nimo aceptado (ej. 88): "))

# Par√°metros de flujo
flujo = 2.0
flujo_min = 0.5
flujo_max = 5.0

# Historial para graficar
hist_spo2 = []
hist_flujo = []
tiempo = []

# Controlador adaptativo
def controlador_adaptativo(spo2_val, flujo_actual):
    if spo2_val < spo2_critico:
        flujo_actual += 0.5
    elif spo2_val < (spo2_critico + 4):
        flujo_actual += 0.2
    else:
        flujo_actual -= 0.3
    return max(flujo_min, min(flujo_actual, flujo_max))

# Funci√≥n de diagn√≥stico simple
def detectar_insuficiencia(historial_spo2):
    bajos = [s for s in historial_spo2 if s < spo2_critico]
    if len(bajos) > (tiempo_simulacion * 0.2):  # m√°s del 20% del tiempo con SpO2 bajo
        return f"‚ö†Ô∏è Sospecha de insuficiencia pulmonar cr√≥nica en {nombre}"
    else:
        return f"‚úÖ {nombre} presenta niveles de SpO‚ÇÇ aceptables"

# Crear figura
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8,6))
line1, = ax1.plot([], [], label="SpO‚ÇÇ (%)", color="blue")
line2, = ax2.plot([], [], label="Flujo O‚ÇÇ (L/min)", color="orange")

def init():
    ax1.set_xlim(0, tiempo_simulacion)
    ax1.set_ylim(80, 100)
    ax1.set_ylabel("SpO‚ÇÇ (%)")
    ax1.axhline(spo2_critico, color='red', linestyle='--', label="SpO‚ÇÇ cr√≠tico")
    ax1.axhline(spo2_critico + 4, color='green', linestyle='--', label="Objetivo")
    ax1.legend()
    ax1.grid()

    ax2.set_xlim(0, tiempo_simulacion)
    ax2.set_ylim(0, 6)
    ax2.set_ylabel("Flujo (L/min)")
    ax2.set_xlabel("Tiempo (s)")
    ax2.legend()
    ax2.grid()

    return line1, line2

def update(frame):
    global spo2, flujo

    # Simular SpO2 con algo de variaci√≥n
    spo2 += random.uniform(-1.5, 1.5)
    spo2 = max(80, min(spo2, 98))

    # Aplicar controlador
    flujo = controlador_adaptativo(spo2, flujo)

    # Guardar historial
    tiempo.append(frame)
    hist_spo2.append(spo2)
    hist_flujo.append(flujo)

    line1.set_data(tiempo, hist_spo2)
    line2.set_data(tiempo, hist_flujo)

    return line1, line2

# Crear animaci√≥n
ani = animation.FuncAnimation(
    fig, update, frames=tiempo_simulacion,
    init_func=init, blit=False, interval=300, repeat=False
)

# Guardar animaci√≥n
nombre_archivo = f"simulacion_{nombre.replace(' ', '_')}.gif"
ani.save(nombre_archivo, writer='pillow')

# Mostrar resultados
display(Image(filename=nombre_archivo))
diagnostico = detectar_insuficiencia(hist_spo2)
print(f"\nüë§ Paciente: {nombre}, {edad} a√±os")
print(f"üìã Diagn√≥stico estimado: {diagnostico}")
```
